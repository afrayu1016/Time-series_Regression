```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data = pd.read_excel('_2021.xls')
data = pd.DataFrame(data)
```

```python
data = data.rename(columns={u"測站                  ": "position", u"日期                  ": "DATE", u"測項   ":"item"})
data.head()
```

```python
data = data.drop('position',axis =1)
data = data.drop(data.index[0])
data.head()
```

### 轉換時間型態

```python
df=data
for index in df.index:
    df.DATE[index] = df.DATE[index].date()
df.head()
```

### 過濾時間範圍 10 月～12月

```python
df_2 = df
df_2['DATE'] = pd.to_datetime(df_2['DATE'], format='%Y-%m-%d')
filtered_df = df.loc[(df_2['DATE'] >= '2021-10-01')
                     & (df_2['DATE'] <= '2021-12-31')]
# Display
filtered_df.reset_index(inplace = True)
filtered_df = filtered_df.drop(['index'],axis =1 )
filtered_df
```

```python
# filter_2 用於處理所有數值
filter_2 = filtered_df.drop(['DATE','item'],axis =1)
filter_2.head()
```

### 處理缺失值

#### 將缺失值做前後一小時的平均處理，若第一個小時為空值則設成0
#### 最後將資料轉成float 型態在和原資料的item及時間做合併


Filter Nan value

```python
#判斷是否為浮點數

def isfloat(num):
    try:
        float(num)
        return True
    except ValueError:
        return False
```

```python
#將缺失值做前後一小時的平均處理，若第一個小時為空值則設成0

for r_index in filter_2.index:

    for i in filter_2.iloc[r_index].iteritems():

            position = i # 紀錄當下的位置

            if isfloat(position[1]): # 判斷是否為數字        
                last = position # last紀錄上一個數字

            else:
                c_index = i[0]

                if c_index == 0: # 若一開始即為空值，則將第一個數值設為0

                    list_position = list(position) # 更新紀錄的position的值
                    list_position[1] = 0
                    position = tuple(list_position)

                    filter_2.iloc[[r_index],[c_index]] = 0 

                else:
                    next_one = 0 # 取消此行，則ｔａｉｌ皆為空的數字為最後一個有被記錄的數字

                    for j in range(c_index +1 ,len(filter_2.iloc[r_index])): # 找到下一個非空值的值
                        if isfloat(filter_2.iloc[[r_index],[j]].values):
                            next_one = filter_2.iloc[[r_index],[j]].values
                            break

                    list_position = list(position)
                    list_last = list(last)
                    list_position[1] = (list_last[1] + next_one)/2.0 
                    position = tuple(list_position)

                    new_value = float(list_position[-1]) # 將新的value取代到原本的dataframe:filter_2
                    filter_2.iloc[[r_index],[c_index]] = new_value

                last = position 
```

將資料型態轉成float

```python
filter_2 = filter_2.astype(float)
filter_2.info()
```

```python
df_pro_nan = pd.concat([filtered_df.DATE,filtered_df.item ,filter_2.reindex(filtered_df.index)],axis =1)
df_pro_nan.head()
```

### 製作時序資料: 將資料形式轉換為行(row)代表18種屬性，欄(column)代表逐時數據資料



```python
item_index = list(dict.fromkeys(df_pro_nan.item))
date_index = list(dict.fromkeys(df_pro_nan.DATE))
```

```python
df_pro_nan.set_index(keys = ["DATE","item"],inplace=True)
df_pro_nan.head()
```

```python
# 製作新的dataframe
date_arr = []
hour_arr =[]

for j in date_index:
    for i in range(0,24):
        date_arr.append(j)
        hour_arr.append(i)
        
arrays = [np.array(date_arr),
            np.array(hour_arr)]
new_format = pd.DataFrame(index=item_index,columns= arrays)
```

```python
import datetime
from datetime import timedelta, date

end_date = df_pro_nan.index[-1][0]

# 利用時間index 來轉換成新的格式, 轉成18 X(92X24) (包含12月的資料)

for index in range(0,len(df_pro_nan.index),18): 
    if index == 0:
        date = df_pro_nan.index[index][0]
        
    else:   
        date = df_pro_nan.index[index-1][0] + datetime.timedelta(days=1)
    if date > end_date:
        break
    new_format [date] = df_pro_nan.loc [date]
    
new_format.head()
```

### 製作PM2.5的時間切割集 ： 取6小時，預測第1&6個小時


```python
period = 7
col = [1,2,3,4,5,6]
model_data = pd.DataFrame(columns = col)

# 切割Ｘ為6個小時為一個特徵
for start_hour in range(0,len(new_format.columns)):
    range_data = pd.DataFrame(new_format.loc['PM2.5               '][start_hour:start_hour+period-1]).T
    try:
        d = pd.DataFrame(range_data.values, columns = col) #若無法執行，表示已超出時間範圍
        model_data = pd.concat([model_data,d],ignore_index = True)
    except:
        break
        
#取得未來第一個小時的pm2.5數值 : length 2203
t = []
for target in range(6,len(new_format.columns)):
    t.append(new_format.loc['PM2.5               '][target])
pm_1 = pd.DataFrame( t, columns = ['PM2.5'])

#取得未來第6個小時的pm2.5數值 : length 2197
t_2 = []
for target in range(11,len(new_format.columns)):
    t_2.append(new_format.loc['PM2.5               '][target])
pm_6 = pd.DataFrame( t_2, columns = ['PM2.5'])

```

### 製作所有特徵的時間切割集 ： 取6小時，預測第1&6個小時


```python
period = 7
model_data_all = pd.DataFrame()
h_range = [1,2,3,4,5,6]

# 把每一個item的時間序列分別做，再做concat
for item_index in new_format.index:
    item_arr = []
    for c in range(1,7):
            item_arr.append(item_index)
            col = [np.array(item_arr),np.array(h_range)]
            
    item_df =pd.DataFrame(columns = col)

    for start_hour in range(0,len(new_format.columns)):
        range_data = pd.DataFrame(new_format.loc[item_index][start_hour:start_hour+period-1]).T
        
        try:
            d = pd.DataFrame(range_data.values, columns = col) #若無法執行，表示已超出時間範圍
            item_df= pd.concat([item_df,d],ignore_index = True)
            
        except:
            break
            
    model_data_all = pd.concat([model_data_all,item_df], axis=1)
```

### Split data to training set (Oct.~Nov.) and testing set (Oct.)

```python
train_period_1 = 61*24-6
train_period_6 = 61*24-11

# predict 1st hour value
pm_train_1 = model_data[:train_period_1]
pm_test_1 = model_data[train_period_1:-1]

all_train_1 = model_data_all[:train_period_1]
all_test_1= model_data_all[train_period_1:-1]

y_train_1 = pm_1[:train_period_1]
y_test_1 = pm_1[train_period_1:]

# predict 6st hour value
pm_train_6 = model_data[:train_period_6]
pm_test_6 = model_data[train_period_6:-6]

all_train_6 = model_data_all[:train_period_6]
all_test_6= model_data_all[train_period_6:-6]

y_train_6 = pm_6[:train_period_6]
y_test_6 = pm_6[train_period_6:]
```
